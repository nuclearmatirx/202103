```
import requests
```

```
# Coronavirus API URL
url = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/F-D0047-003?Authorization=CWB-D29B98CC-2679-4D78-8AA8-006A14A6C095&format=JSON&locationName=%E7%A4%81%E6%BA%AA%E9%84%89'
```

```
def retrieval_api(api_url):
  response = requests.get(api_url)
  # print('Status Code: ' + str(response.status_code))
  return str(response.json())
```

```
location_json = retrieval_api(url)
```

```
print(location_json)
```

```
print(type(location_json))
```

```
location_json = location_json.replace("\'", "\"")
print(location_json)
```
